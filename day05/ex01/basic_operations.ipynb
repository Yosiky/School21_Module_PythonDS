```python
import pandas as pd
```

```python
if __name__ == "__main__":
    #################################
    print("#"*100)
    data = pd.read_csv("../feed-views.log", header=None, names=["datetime", "user"], sep='\t', parse_dates=["datetime"])
    data["year"] = data["datetime"].dt.year
    data["month"] = data["datetime"].dt.month
    data["day"] = data["datetime"].dt.day
    data["hour"] = data["datetime"].dt.hour
    data["minute"] = data["datetime"].dt.minute
    data["second"] = data["datetime"].dt.second
    print(data)
    #################################
    print("#"*100)
    data["daytime"] = pd.cut(data["hour"], [0, 4, 7, 11, 17, 20, 24], right=False, include_lowest=True,
                          labels=["night", "early morning", "morning", "afternoon", "early evening", "evening"])
    data.set_index("user", inplace=True)
    print(data)
    #################################
    print("#"*100)
    print(data.count(), end="\n\n")
    print(data["daytime"].value_counts())
    #################################
    print("#"*100)
    data.sort_values(by=["hour", "minute", "second"], inplace=True)
    print(data)
    #################################
    print("#"*100)
    print("Night max: ", data[data["daytime"] == "night"]["hour"].max())
    print("Morinig min: ", data[data["daytime"] == "morning"]["hour"].min())
    print("Users: ", set(data[data[data["daytime"] == "night"]["hour"].max() == data["hour"]].index))
    print("Mode hour: ", data["hour"].mode())
    print("Mode daytime: ", data["daytime"].mode())
    #################################
    print("#"*100)
    print("nsmallest:")
    print(data[data["daytime"] == "morning"].nsmallest(3, ['hour', 'minute', 'second']))
    print("nlargest:")
    print(data[data["daytime"] == "morning"].nlargest(3, ['hour', 'minute', 'second']))
    #################################
    print("#"*100)
    print(data.describe())
    irq = data.describe()["hour"]["75%"] - data.describe()["hour"]["25%"]
    print("Irq =", irq)
```
