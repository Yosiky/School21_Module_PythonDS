```python
import pandas as pd
```

```python
def get_model(str):
    line = str.split(' ')
    if len(line) > 1:
        return line[1]
    return None
```

```python
if __name__ == "__main__":
    #################################
    print("#"*100)
    data = pd.read_csv("../auto.csv", sep=',', index_col="ID")
    print(data)
    #################################
    print("#"*100)
    print(data.count())
    #################################
    print("#"*100)
    data.drop_duplicates(["CarNumber", "Make_n_model", "Fines"], keep="last", inplace=True)
    print(data.count())
    #################################
    print("#"*100)
    print("Count NaN:")
    print(data.isna().sum())
    data.dropna(axis='columns', thresh=500, inplace=True)
    print("\nCount NaN:")
    print(data.isna().sum())
    #################################
    print("#"*100)
    data["Refund"].fillna(method="ffill", inplace=True)
    print("Count NaN:")
    print(data.isna().sum())
    #################################
    print("#"*100)
    data["Fines"].fillna(data["Fines"].mean(), inplace=True)
    print("Count NaN:")
    print(data.isna().sum())
    #################################
    print("#"*100)
    data["Make"] = data["Make_n_model"].apply(lambda x : x.split(' ')[0])
    data["Model"] = data["Make_n_model"].apply(get_model)
    data.drop(columns="Make_n_model", inplace=True)
    print(data)
    data.to_json("auto.json", orient="records")
```
