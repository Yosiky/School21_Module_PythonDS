```python
import pandas as pd
import numpy as np
import requests

np.random.seed(21)
pd.options.display.float_format = '{:,.2f}'.format
```

# Read JSON file

```python
data = pd.read_json("../ex02/auto.json")
data.info()
data
```

# Create sample

```python
sample = data.sample(200, random_state=21)
sample
```

```python
new_data = pd.concat([data, sample])
new_data
```

# Create column Year

```python
np.random.seed(21)
Year = pd.Series([np.random.randint(1980, 2020) for i in range(len(new_data))], name='Year')
fines = new_data
fines['Year'] = Year
fines.head()
```

# Read second file "surname.json"

```python
df2 = pd.read_json("../surname.json")
df2.columns = df2.iloc[0]
df2.drop(0, inplace=True)
df2
```

# Create surnames

```python
surname = df2["NAME"].sample(len(new_data), random_state=21, replace=True)
surname
```

```python
df = pd.DataFrame({"CarNumber" : new_data["CarNumber"].reset_index(drop=True), "SURNAME" : surname.reset_index(drop=True)})
df
```

```python
def create_car_number():
    arr = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
    arr = np.concatenate([arr, range(0, 10)])
    return np.random.choice(arr) + str(np.random.randint(100, 999)) + np.random.choice(arr) + np.random.choice(arr) + "21RUS"
```

```python
new_fines = pd.DataFrame({"CarNumber": [create_car_number() for i in np.arange(5)]})
new_fines["Refund"] = pd.Series([np.random.choice(new_data["Refund"]) for i in np.arange(5)])
new_fines["Fines"] = pd.Series([np.random.choice(new_data["Fines"]) for i in np.arange(5)])
new_fines["Make"] = pd.Series([np.random.choice(new_data["Make"]) for i in np.arange(5)])
new_fines["Model"] = pd.Series([np.random.choice(new_data["Model"]) for i in np.arange(5)])
new_fines["Year"] = pd.Series([np.random.randint(1980, 2019) for i in np.arange(5)])
new_fines
```

```python
fines = pd.concat([fines, new_fines])
fines
```

```python
df = df[:-20]
add_new_3_line = pd.DataFrame({"CarNumber" : [create_car_number() for i in range(3)],
                              'SURNAME': [np.random.choice(df['SURNAME']) for i in range(3)]})
df = pd.concat([df, add_new_3_line])
df
```

```python
fines
```

```python
res = fines.merge(df, 'inner', on="CarNumber")
res
```

```python
pd.pivot_table(res, values="Fines", index=['Make', 'Model'], columns=['Year'], aggfunc=np.sum)
```
